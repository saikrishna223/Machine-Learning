```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('Wine_data.csv')
```

```python
df.head()
```

```python
df['quality'].unique()
```

# Creating bins

```python
#Define bins
bins = [0,6,10]
#Name of group
group_names = ['Bad','Good']
df['Quality']=pd.cut(df['quality'],bins,labels = group_names)
df.head()
```

# Using One Hot Encoding

```python
df1=pd.get_dummies(df,drop_first=True)
```

```python
df1.head()
```

# Dropping Unwanted Fields

```python
df2=df1.drop(['quality'],axis=1)
```

```python
df2.head()
```

# Assigning X and y values

```python
X=df2.iloc[ :, :11]
```

```python
y=df2.iloc[ :, -1:]
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Scaling 

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100,max_depth=1700,random_state=42)
rf.fit(X_train, y_train)
predict = rf.predict(X_test)

from sklearn import metrics
print('Accuracy :: ',metrics.accuracy_score(y_test,predict))
```

# Using GridSearchCV 

```python
from sklearn.model_selection import GridSearchCV
model1 = RandomForestClassifier(n_estimators=50, max_depth=10, max_features='log2')

param_grid = {
    'n_estimators' : [50,1000],
    'max_depth': [1,100],
    'max_features': ['auto', 'sqrt', 'log2'],
    'random_state':[0,1000]
}


CV_rfc = GridSearchCV(estimator=model1, param_grid=param_grid, cv= 5)
CV_rfc.fit(X, y)
print (CV_rfc.best_params_)
```

# Assigning best parameters

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50,max_depth=100,random_state=0,max_features='auto')
rf.fit(X_train, y_train)
predict = rf.predict(X_test)

from sklearn import metrics
print('Accuracy :: ',metrics.accuracy_score(y_test,predict))
```

# ExtraTreesClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier
et = ExtraTreesClassifier( criterion='entropy', max_depth=110, max_features='log2',
                     min_samples_leaf=2, min_samples_split=5,
                     n_estimators=125,
                     n_jobs=None,random_state=5374)

et.fit(X_train, y_train)
predict = et.predict(X_test)

from sklearn import metrics
print('Accuracy :: ',metrics.accuracy_score(y_test,predict))
```
